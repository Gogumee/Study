```python
def euclidean_norm(x1, x2):
    #print(np.sqrt(((x1 - x2)**2).sum(axis=0)))
    #return np.sqrt((x1 - x2)**2)
    return np.sqrt(((x1 - x2)**2).sum(axis=0))
def _call_norm(x1, x2):
    if len(x1.shape) == 1:
        x1 = x1[np.newaxis, :]
    if len(x2.shape) == 1:
        x2 = x2[np.newaxis, :]
    x1 = x1[..., :, np.newaxis]
    x2 = x2[..., np.newaxis, :]
    #print('x1.shape : ', x1.shape)
    #print('x2.shape : ', x2.shape)
    #print('euclidean_norm.shape : ', euclidean_norm(x1, x2))
    return euclidean_norm(x1, x2)
def _h_linear(r):
    return r
def A(xi):
    # this only exists for backwards compatibility: self.A was available
    # and, at least technically, public.
    r = _call_norm(xi, xi)
    return self._init_function(r) - np.eye(self.N)*self.smooth
```

```python
def rbf_test(*args):
    xi = np.asarray([np.asarray(a, dtype=np.float_).flatten()
                       for a in args[:-1]])
    N = xi.shape[-1]
    di = np.asarray(args[-1]).flatten()

    print('xi.shape : ', xi.shape)
    print('N : ', N)
    print('di : ', di.shape)

    for x in xi :
        print('x.size : ', x.size)
    print('di.size : ', di.size)
    if not all([x.size == di.size for x in xi]):
        raise ValueError("All arrays must be equal length.")


    norm = euclidean_norm
    epsilon = 2
    smooth = 1
    function = _h_linear

    r = _call_norm(xi, xi)
    A = function(r) - np.eye(N)*smooth
    print('A.shape : ', A.shape)
    
    nodes = scipy.linalg.solve(A, di)
    print('nodes.shape : ', nodes.shape)
    
    g_y, g_x = np.mgrid[0:15:16j, 0:15:16j]
    shp = g_x.shape
    xa = np.asarray([a.flatten() for a in [g_x, g_y]], dtype=np.float_)
    print('xa.shape : ', xa.shape)
    r2 = _call_norm(xa, xi)
    print('r2.shape : ', r2.shape)
    return np.dot(function(r2), nodes).reshape(shp)
```

```python
def euclidean_norm_tf(x1, x2):
    #return tf.sqrt(tf.square(tf.subtract(x1, x2)))
    return tf.reduce_sum(tf.sqrt(tf.square(tf.subtract(x1, x2))), 1) 
    #np.sqrt(((x1 - x2)**2).sum(axis=0))

def _h_linear_tf(r):
    return r

def _call_norm_tf(x1, x2):
    x1 = tf.expand_dims(x1, 3) 
    x2 = tf.expand_dims(x2, 2) 
    print('x1.shape : ', x1.shape)
    print('x2.shape : ', x2.shape)
    print('euclidean_norm.shape : ', euclidean_norm_tf(x1, x2).shape)

    return euclidean_norm_tf(x1, x2)

# def A(xi):
#     # this only exists for backwards compatibility: self.A was available
#     # and, at least technically, public.
#     r = _call_norm(xi, xi)
#     return self._init_function(r) - np.eye(self.N)*self.smooth

def rbf_tf(n_x, n_y, x, xa, batch_shape) :
    xi = tf.concat([tf.reshape(n_x, (-1, 1, t_l_num)), tf.reshape(n_y, (-1, 1, t_l_num))], 1)
    N = xi.shape[-1].value
    di = tf.expand_dims(x, 2)  #tf.reshape(x, ())
    print('xi.shape', xi.shape)
    
    norm = euclidean_norm_tf
    epsilon = tf.constant(2.)
    smooth = tf.constant(1.)
    
    function = _h_linear_tf
    
    r = _call_norm_tf(xi, xi)
    A = tf.subtract(r, tf.multiply(smooth, tf.eye(N, batch_shape= batch_shape)))
    print('A.shape : ', A.shape)
    print('di.shape', di.shape)

    nodes = tf.matrix_solve (A, di)
    print('nodes.shape : ', nodes.shape)
    
    print('xa.shape', xa.shape)
    print('xi.shape', xi.shape)

    r2 = _call_norm_tf(xa, xi)
    print('r2.shape', r2.shape)
    #return tf.matmul(r2, nodes)
    return tf.reshape(tf.matmul(r2, nodes), [-1, t_g_shape[0], t_g_shape[1]]) 
```

```python
t = tf.Graph()

with t.as_default():
    a_x_input = tf.placeholder(tf.float32, (None, t_l_num), name= 'a_x_input')
    a_y_input = tf.placeholder(tf.float32, (None, t_l_num), name= 'a_y_input')
    b_x_input = tf.placeholder(tf.float32, (None, t_l_num), name= 'b_x_input')
    b_y_input = tf.placeholder(tf.float32, (None, t_l_num), name= 'b_y_input')
    
    #g_x_input = tf.placeholder(tf.float32, (None, t_g_num, t_g_num))
    #g_y_input = tf.placeholder(tf.float32, (None, t_g_num, t_g_num))
    
    i_input = tf.constant(image, dtype=tf.float32, name= 'i_input')
    g_input = tf.constant(t_xa, dtype=tf.float32, name= 'g_input')
    
    batch_shape = tf.placeholder(tf.int32, [1], name= 'batch_shape')
    
    rbf_x_logits = rbf_tf(a_x_input, a_y_input, b_x_input, g_input, batch_shape)
    rbf_y_logits = rbf_tf(a_x_input, a_y_input, b_y_input, g_input, batch_shape)
    
    #warped = cv2.remap(image, map_x, map_y, cv2.INTER_LINEAR)
    with tf.Session() as sess :
        sess.run(tf.global_variables_initializer())
        
        feed_dict = {
            a_x_input : a_x.reshape(-1, t_l_num),
            a_y_input : a_y.reshape(-1, t_l_num),
            b_x_input : b_x.reshape(-1, t_l_num),
            b_y_input : b_x.reshape(-1, t_l_num),
            #g_x_input : g_x.reshape(-1, t_g_num, t_g_num),
            #g_y_input : g_y.reshape(-1, t_g_num, t_g_num),
            batch_shape : [1]
        }
        
        x, y = sess.run([rbf_x_logits, rbf_y_logits], feed_dict= feed_dict)
        print(x.shape, y.shape)
```
